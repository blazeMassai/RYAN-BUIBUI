```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
htm=urlopen('http://www.pythonscraping.com/pages/page1.html')
bs=BeautifulSoup(htm.read(), 'html.parser')
nameList=bs.find_all('span', {'class':'red'})
for name in nameList:
    print(name.get_text())
```

```python
ht2=urlopen('https://www.trc.co.tz')
bs1=BeautifulSoup(ht2.read(), 'html5lib')
tagList=bs1.find_all('h1', {'class':'title-decoration'})
for tag in tagList:
    print(tag.get_text())
```

```python
ht3=urlopen('http://ajira.go.tz')
bs2=BeautifulSoup(ht3.read(), 'html.parser')
ajiras=bs2.find_all(['h1','h2','h3','h4','h5','h6'])
for ajira in ajiras:
    print(ajira.get_text())
```

```python
ht4=urlopen('http://ajira.go.tz')
bs3=BeautifulSoup(ht4.read(), 'html.parser')
kazis=bs3.find_all('h3',{'class':'advert_news_title'})
for kazi in kazis:
    print(kazi.get_text())
```

```python
ht5=urlopen('https://www.ega.go.tz')
bs4=BeautifulSoup(ht5.read(), 'html.parser')
majina = bs4.find_all(id='navbarDropdown', class_='nav-link')
for jina in majina:
    print(jina.get_text())
```
